# **Training And Testing Available Data**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/6_train_test_split/carprices.csv")
display(df.head())
```

```python
sns.scatterplot(x="Mileage",y="Sell Price($)",data=df,color="r")
plt.show()
```

```python
sns.scatterplot(x="Age(yrs)",y="Sell Price($)",data=df,color="r")
plt.show()
```

```python
X=df[["Mileage"	,"Age(yrs)"]]
X
```

```python
Y=df["Sell Price($)"]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
reg.predict(X_test)
```

```python
Y_test
```

```python
reg.score(X_test,Y_test)
```
