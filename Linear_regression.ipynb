# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_frame=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/homeprices.csv")
```

```python
data_frame
```

```python
plt.scatter(data_frame.area,data_frame.price,color="g")
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price Plot")
plt.show()
```

```python
new_data=data_frame.drop("price",axis=1)
```

```python
new_data
```

```python
price=data_frame["price"]
price
```

```python
from sklearn import linear_model
```

```python
# create linear regression
reg=linear_model.LinearRegression()
reg.fit(new_data,price)
```

Find the price of 3300 sqft area

```python
reg.predict([[3300]])
```

SLOPE

```python
reg.coef_
```

INTERCEPT

```python
reg.intercept_
```

PRICE = ( SLOPE * AREA ) + INTERCEPT

```python
135.78767123*3300+180616.43835616432
```

Find the price of the area of 5000 sqft

```python
reg.predict([[5000]])
```

find the price of a list of areas

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/areas.csv")
```

```python
df
```

```python
prices_list=reg.predict(df)
prices_list
```

```python
df["price"]=prices_list
```

```python
df
```

```python
df.to_csv("price_prediction")
```
