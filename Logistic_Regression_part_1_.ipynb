**Logistic Regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/7_logistic_reg/insurance_data.csv")
```

```python
df
```

```python
sns.scatterplot(x="age",y="bought_insurance",data=df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df[['age']],df.bought_insurance)
```

```python
X_train
```

```python
len(X_train)
```

```python
X_test
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
nw=model.predict(X_test)
```

```python
Y_test
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
df_test=X_test
df_test
```

```python
df_test["bought_insurance"]=Y_test
```

```python
df_test["bought_insurance(predicted)"]=nw
```
