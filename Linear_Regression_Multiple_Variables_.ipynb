# Linear Regression Multiple Variables

```python
!pip install word2number
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import math
from word2number import w2n
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/homeprices.csv")
```

```python
df.head()
```

```python
df["bedrooms"].median()
```

```python
df.bedrooms.fillna(df["bedrooms"].median(),inplace=True)
```

```python
df.head()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df.drop("price",axis=1),df.price)
```

**find the price of 3000 sqr ft area, 3 bedrooms, 40 year old**

```python
reg.predict([[3000, 3, 40]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
112.0624419*3000+23388.88007794*3+-3231.71790863*40+221323.00186540396
```

**find the price of 2500 sqr ft area, 4 bedrooms, 5 year old**

```python
reg.predict([[2500,4,5]])
```

# Exercise

In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

Answer
53713.86 and 93747.79

```python
df1=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/Exercise/hiring.csv")
```

```python
df1.head(10)
```

```python
df1.isnull().sum()
```

```python
df1.experience.fillna("zero",inplace=True)
```

```python
df1
```

```python
df1["test_score(out of 10)"].fillna(df1["test_score(out of 10)"].median(),inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1.head()
```

```python
df1["experience"]=df1["experience"].apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)
```

```python
reg1=linear_model.LinearRegression()
reg1.fit(df1[['experience','test_score(out of 10)','interview_score(out of 10)']],df1["salary($)"])
```

```python
reg1.predict([[2,9,6]])
```

```python
reg1.predict([[12,10,10]])
```
