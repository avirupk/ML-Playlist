```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
dfl=df
dfl["Car Model"]=le.fit_transform(df["Car Model"])
```

```python
dfl
```

```python
X=dfl[["Car Model","Mileage","Age(yrs)"]].values
X
```

```python
Y=dfl["Sell Price($)"].values
Y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ohe = OneHotEncoder()
cl=ColumnTransformer([("Car Model",ohe,[0])],remainder="passthrough")
```

```python
X=cl.fit_transform(X)
X
```

```python
X=X[:,1:]
X
```

```python
from sklearn import linear_model
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(X,Y)
```

```python
# 0 0  -> Audi A5
# 0 1  -> Mercedez Benz C class
# 1 0  -> BMW X5
```

 **Predict price of a mercedez benz that is 4 yr old with mileage 45000**

```python
reg.predict([[0,1,45000,4]])
```

**Predict price of a BMW X5 that is 7 yr old with mileage 86000**

```python
reg.predict([[1,0,86000,7]])
```

**Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())**

```python
reg.score(X,Y)
```
