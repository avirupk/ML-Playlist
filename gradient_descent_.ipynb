# **Gradient Descent and Cost Function**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def gradient_descent(x,y):
  m_curr=b_curr=0
  learning_rate=0.01
  iteration=1000
  n=len(x)
  for i in range(iteration):
    y_predicted=m_curr*x+b_curr
    # plt.plot(x,y_predicted,"c")
    mse=(1/n)*sum([val**2 for val in (y-y_predicted)])
    dm=-(2/n)*sum(x*(y-y_predicted))
    db=-(2/n)*sum(y-y_predicted)
    m_curr=m_curr-learning_rate*dm
    b_curr=b_curr-learning_rate*db
    print(f"MSE = {mse},\tm = {m_curr},\tb = {b_curr},\titeration = {i}")

```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```

```python
import math
```

```python
def gradient_descent1(math_x,cs_y):
  m_curr=b_curr=0
  learning_rate=0.0002
  iteration=1000000
  n=len(math_x)
  previous_mse=0
  for i in range(iteration):
    cs_y_predicted=m_curr*math_x+b_curr
    mse=(1/n)*sum([val**2 for val in (cs_y-cs_y_predicted)])
    dm=-(2/n)*sum(math_x*(cs_y-cs_y_predicted))
    db=-(2/n)*sum(cs_y-cs_y_predicted)
    m_curr=m_curr-learning_rate*dm
    b_curr=b_curr-learning_rate*db
    if math.isclose(mse,previous_mse,rel_tol=1e-20):
      break
    previous_mse=mse
    print(f"MSE = {mse},\tm = {m_curr},\tb = {b_curr},\titeration = {i}")

  return m_curr,b_curr
```

```python
p=reg.coef_
q=reg.intercept_
```
